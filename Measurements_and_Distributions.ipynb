# Introduction

## The Scientific Method

Many of you are familiar with the scientific method:

*   You have a question you're interested in
*   You come up with a explanation (a hypothesis) that answers that question
*   You design an experiment to test that explanation by making a measurement
*   You analyze your data to see if they support your explanation
*   If the data don't support your explanation, you repeat this process with a revised explanation.


We only know stuff because we run experiments. These experiments aren't restricted to the lab! We *constantly* run little experiments all day - we make decisions about how to act based on what we detect in our environment:

*   we try different foods to see what feels most satisfying
*   we try different apps to see if anything will help with our productivity
*   when we detect we're too cold we layer up

**In physics and other sciences, we conduct formal experiments to answer questions we care about.
```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
plt.rcParams['font.size'] = '16'
from IPython import display


def plot_circle(ax,center,radius,color):
    theta = np.linspace(0,2*np.pi,200)
    xx = radius*np.cos(theta) + center[0]
    yy = radius*np.sin(theta) + center[1]
    ax.plot(xx,yy,color=color,lw=1)

def make_dart_plots(xx,yy,colormap):

    # create the plot
    fig,ax = plt.subplots(1,2,figsize=(16,7),sharey=True)

    # scatterplot
    h = ax[0].plot(xx,yy,'rX',markersize=15,alpha = 0.2)
    datarange = np.amax(abs(mydata))
    edges = np.arange(-datarange-0.5,datarange+1,1)
    datarange = np.amax(abs(mydata))
    edges = np.arange(-datarange-0.5,datarange+1,1)

    # histogram
    h = ax[1].hist2d(xx,yy,bins=edges)
    
    n = int(np.amax(h[0])) + 1
    cmap = plt.get_cmap(colormap,n)
    h = ax[1].hist2d(xx,yy,bins=edges,cmap=cmap)    
    divider = make_axes_locatable(ax[1])
    cax = divider.append_axes("right",size="5%",pad=0.05)

    L = n-1
    delta = L/2/n


    tick_locs = np.arange(delta,L-delta+1,2*delta)
    cbar = fig.colorbar(h[3],cax=cax,ticks=tick_locs)
    cbar.set_label('num of hits')
    tick_labels = np.arange(n)
    cbar.set_ticklabels(np.arange(n))


    # label the plots
    rr = [0.5, 1.5, 4, 5, 7.6, 8.6]
    ax[0].set_ylabel('y coordinate');
    for axis in ax:
        for r in rr:
            if axis == ax[0]:
                color = 'k'
            else:
                color = 'k'
            plot_circle(axis,(0,0),r,color)
            
        axis.set_aspect(1)
        axis.set_xlim(-datarange-0.5,datarange+0.5)
        axis.set_ylim(-datarange-0.5,datarange+0.5)
        axis.set_xlabel('x coordinate')
        axis.set_xticks(range(-datarange,datarange+1))
        axis.set_yticks(range(-datarange,datarange+1))
        axis.tick_params(axis='both', which='major', labelsize=10)
```
